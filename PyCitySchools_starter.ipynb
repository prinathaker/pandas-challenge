```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])

def format_float(value):
  return f'{value:,.2f}'
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create School Summary

# no of students in each school
student_by_school = school_data_complete['school_name'].value_counts()
student_by_school

# School Type
school_type = school_data_complete.groupby('school_name')['type'].unique()

school_type = school_type.str[0]

school_type

# total budget for each school
school_budget = school_data.groupby('school_name')['budget'].sum()
school_budget

# budget for each student
#student_budget = school_data_complete.groupby('school_name')['budget'].sum()/student_by_school
student_budget = school_budget/student_by_school

student_budget.head()

# average math score for each school
school_avg_math = round(school_data_complete.groupby('school_name')['math_score'].mean(),6)

school_avg_math.head()

# average reading score for each school
school_avg_read = round(school_data_complete.groupby('school_name')['reading_score'].mean(),6)

school_avg_read.head()

#passing math with greater then 70
passing_math_student = school_data_complete.loc[school_data_complete['math_score']>=70.0]
passing_math_student  
per_passing_math = round((passing_math_student.groupby('school_name')['math_score'].count()/student_by_school)*100,6)
per_passing_math

#passing reading with greater then 70
passing_reading_student = school_data_complete.loc[school_data_complete['reading_score']>=70.0]
passing_reading_student  
per_passing_reading = round((passing_reading_student.groupby('school_name')['reading_score'].count()/student_by_school)*100,6)
per_passing_reading

#Overall passing Rate
overall_passing = round((per_passing_math+per_passing_reading)/2,6)
overall_passing

# Create Data frame to hold School summary results
school_data_summary_df = pd.DataFrame({'School Type':school_type,
                                        'Total Students' : student_by_school,
                                        'Total School Budget' :school_budget,
                                       'Budget Per Student' :student_budget, 
                                       "Average Math Score" :school_avg_math,
                                        "Average Reading Score" :school_avg_read,
                                        "Percent Passing Math" :per_passing_math,
                                        "Percent Passing Reading" :per_passing_reading,
                                        "Percent Overall Passing Rate" :overall_passing}
                          ,columns = ["School Type","Total Students","Total School Budget","Budget Per Student",
                                                "Average Math Score","Average Reading Score","Percent Passing Math",
                                                "Percent Passing Reading","Percent Overall Passing Rate"])

#school_data_summary_df["Total Students"] = school_data_summary_df["Total Students"].map("{:,}".format)
school_data_summary_df["Total School Budget"] = school_data_summary_df["Total School Budget"].map("${:,.2f}".format)
school_data_summary_df["Budget Per Student"] = school_data_summary_df["Budget Per Student"].map("{:,.2f}".format)
school_data_summary_df
```
